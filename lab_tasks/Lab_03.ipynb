```python
import math

def knn(X_train, y_train, X_test, k):
    y_pred = []
    for i in range(len(X_test)):
        distances = []
        for j in range(len(X_train)):
            dist = math.sqrt(sum([(X_test[i][x] - X_train[j][x])**2 for x in range(len(X_test[i]))]))
            distances.append((dist, y_train[j]))
        distances.sort()
        neighbors = distances[:k]
        counts = {}
        for n in neighbors:
            if n[1] in counts:
                counts[n[1]] += 1
            else:
                counts[n[1]] = 1
        y_pred.append(max(counts, key=counts.get))
    return y_pred
```

```python
import math

def euclidean_distance(x1, x2):
    """
    Calculate the Euclidean distance between two points
    """
    distance = 0
    for i in range(len(x1)):
        distance += (x1[i] - x2[i]) ** 2
    return math.sqrt(distance)

def get_neighbors(X_train, y_train, x_test, k):
    """
    Find the k nearest neighbors of a test point
    """
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(X_train[i], x_test)
        distances.append((i, dist))
    distances.sort(key=lambda x: x[1])
    neighbors = []
    for i in range(k):
        neighbors.append(y_train[distances[i][0]])
    return neighbors

def predict(X_train, y_train, x_test, k):
    """
    Predict the class label of a test point using KNN
    """
    neighbors = get_neighbors(X_train, y_train, x_test, k)
    counts = {}
    for neighbor in neighbors:
        if neighbor in counts:
            counts[neighbor] += 1
        else:
            counts[neighbor] = 1
    return max(counts, key=counts.get)
```

```python
import pandas as pd
df = pd.read_csv('fruit_data_with_colors _1_.csv')
df = df.drop('fruit_name', axis=1)
if df.isnull().values.any():
    df = df.fillna(df.mean())
train_data = df.iloc[:50, :]
test_data = df.iloc[50:, :]
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train = train_data.drop('fruit_label', axis=1)
y_train = train_data['fruit_label']
X_test = test_data.drop('fruit_label', axis=1)
y_test = test_data['fruit_label']

k_values = range(1, 11)
accuracy_scores = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print(f'K = {k}, accuracy = {accuracy:.2f}')
import matplotlib.pyplot as plt

plt.plot(k_values, accuracy_scores, marker='o')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('Accuracy scores for different values of K')
plt.show()
```
